```python
import cv2
import numpy as np
import glob
import os
```

```python
no_seq = 45 #number of videos for each class
frames = 60
```

```python
DATA_PATH = os.path.join('video_data') 
actions = np.array(['All',
'Boy',
'Girl',
'Book',
'Mobile',
'Car',
'Baby',
'Door',
'End',
'Eat',
'Drink',
'Father',
'Mother',
'Go',
'Good',
'Bad',
'Food',
'House',
'In',
'Out',
'Sad',
'Happy',
'Man',
'Woman',
'Stop',
'School',
'New',
'Old',
'Play',
'Room',
'See',
'Sit',
'Sister',
'Brother',
'Think',
'Work',
'name',
'Yes',
'No',
'Walk',
'Love',
'Need',
'Respect',
'Money',
'I',
'You',
'Day',
'Ambulance',
'Buy',
'Bread',
])
```

```python
# Create DataPaths for Videos
for action in actions: 
    for sequence in range(no_seq):
        try: 
            os.makedirs(os.path.join(DATA_PATH, action))
        except:
            print('failed')
```

```python
## Code
for label in actions:
    for vid_no in range(45):
        # get all Filenames of Frames
        path = '/'.join([label,str(vid_no),'*.jpg'])
        files = sorted(glob.glob(path), key= lambda x : int(x.split("\\")[1][:-4]) )
        
        # load all Frame in This Array
        img_array = []
        for filename in files:
            img = cv2.imread(filename)
            height, width, layers = img.shape
            size = (width,height)
            img_array.append(img)
        #Output Path
        vid_path = '/'.join(['./Dataset',label,str(vid_no)+'.mp4'])
        out = cv2.VideoWriter(vid_path,cv2.VideoWriter_fourcc(*'DIVX'), 30, size)
         
        # Write Video
        for i in range(len(img_array)):
            out.write(img_array[i])
        out.release()
        
    
```
